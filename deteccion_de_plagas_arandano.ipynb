```python
from ipywidgets import interact, widgets
from IPython.display import display
```

```python
print("APP TEST")
print("--------")
text = widgets.Text(
    placeholder='Ingrese nombre de la imagen',
    description='Nombre...',
    disabled=False
)
display(text)
```

```python
def callback(wdgt):
    uploaded_file = wdgt.value
    widgets.Image(value=uploaded_file.content.tobytes())
```

```python
uploader = widgets.FileUpload(
    accept='',
    multiple=False
)
#display(uploader)
#uploader.on_submit(callback)


label = widgets.Label("")
ulimg = widgets.Image(value=b'')

def on_upload_change(change):
    label.value = "Loaded"
    uploaded_file = uploader.value
    ulimg.value = value=uploaded_file[0].content.tobytes()

uploader.observe(on_upload_change, names='value')

display(uploader, label, ulimg)
```
